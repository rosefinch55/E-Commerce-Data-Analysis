```python
#!pip install chardet  
import pandas as pd
import chardet

# 检测文件编码（核心代码）
file_path = r'..\data\双十一淘宝美妆数据.csv'

# 读取文件二进制内容（用rb)
with open(file_path, 'rb') as f:
    # 读取前10000字节（足够检测编码，不用读整个文件）
    raw_data = f.read(10000)
    # 检测编码，返回最大概率结果
    detect_result = chardet.detect(raw_data)

# 输出检测结果
print("=== 编码检测结果 ===")
print(f"最大概率编码：{detect_result['encoding']}")
print(f"可信度（0-1）：{detect_result['confidence']:.2f}")
```
